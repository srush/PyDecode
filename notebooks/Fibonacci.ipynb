# Tutorial 1: Fibonacci

```python
import pydecode
import numpy as np
```

We begin by looking at one of the simplest dynamic programming algorithm, generating a fibonacci sequence. 
While there are many ways to generate this sequence, the undelying dynamic programming  algorithm acts as a nice
base for introducing the central data structures in PyDecode.

```python
n = 10
chart = np.zeros(10)
chart[0] = 0
chart[1] = chart[0] + 1
for item in range(2, n):
    chart[item] = chart[item-1] + chart[item-2]
chart
```

```python
chart = pydecode.ChartBuilder(np.arange(10))
chart.init(0)
chart.set(1, [[0]], labels=[0])

for item in range(2, n):
    chart.set(item, [[item-1, item-2]])
graph = chart.finish()
```

```python
weights = pydecode.transform(graph, np.array([1.0]))
inside = pydecode.inside(graph, weights)
inside
```

```python
pydecode.draw(graph, np.array(["+"]*10), graph.node_labeling)
```

```python
pydecode.draw(graph, vertex_labels=np.array(inside, dtype=np.int32))
```
